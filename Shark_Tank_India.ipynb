```python
import pandas as pd

df = pd.read_csv("Shark Tank India.csv")
df.head()
df.info()
df.columns
```

```python
df.isnull().sum()
```

```python
selected_cols = [
    'Startup Name','Industry','Pitchers City','Pitchers State',
    'Original Ask Amount','Original Offered Equity',
    'Total Deal Amount','Total Deal Equity','Number of Sharks in Deal',
    'Namita Investment Amount','Aman Investment Amount','Anupam Investment Amount',
    'Vineeta Investment Amount','Peyush Investment Amount','Ritesh Investment Amount',
    'Amit Investment Amount','Guest Investment Amount'
]

df_small = df[selected_cols]

df_small.to_csv("SharkTank_Clean.csv", index=False)

df_small.head()
```

```python
num_cols = [
    'Original Ask Amount','Original Offered Equity',
    'Total Deal Amount','Total Deal Equity','Number of Sharks in Deal',
    'Namita Investment Amount','Aman Investment Amount','Anupam Investment Amount',
    'Vineeta Investment Amount','Peyush Investment Amount','Ritesh Investment Amount',
    'Amit Investment Amount','Guest Investment Amount'
]

df_small[num_cols] = df_small[num_cols].fillna(0)

cat_cols = ['Industry','Pitchers City','Pitchers State']
df_small[cat_cols] = df_small[cat_cols].fillna("Unknown")

df_small.to_csv("SharkTank_Clean.csv", index=False)

df_small.head()
```
